```python
## Matplot Lib
import matplotlib.pyplot as plt
```

```python
## Plot a linear Graph
plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3],)
plt.grid()
plt.show()
```

```python
## plot a graph between Distance and Time
dist = list(range(2,12,2))# total values 9
time = list(range(1,6)) # total values 5
plt.title("Distance vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,dist,'r-')#(x-axis,y-axis,design)
plt.grid()
plt.show()
```

```python
## Non-linear graph for the v=a*t //velocity = time*acceleration
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# Voltage and Time
import numpy as np
t = np.arange(0.1,2.0,0.1)
v =[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("Time vs Voltage")
plt.grid()

plt.savefig("Voltage-Time.png")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("Income.csv")
df
```

```python
years= df.columns[2:]
incomeAlaska = df.values[1,2:]
plt.title("Income of Alaska")
plt.xlabel("years")
plt.ylabel("Income")
plt.grid()
plt.plot(years,incomeAlaska,'g*')
```

```python
incomelist = []
for i in df.columns[2:]:
    incomelist.append(df[i].mean())
plt.plot(years,incomelist)
plt.title("All States Income")
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.show()
```

## Matplot lib
### Types of plots
    . Bar Graph
    . Histogram
    . Scatter Plot
    . Area Plot
    . Pie Chart

```python
## Example for bar graph
plt.bar([0.50,1.50,2.50,3.50,4.50],[50,40,70,80,20],label="Ferrari",color='g',width=0.5)
plt.bar([0.80,1.80,2.80,3.80,4.80],[80,20,30,70,50],label="Audi",color = 'r',width=0.5)
plt.xlabel("Days")
plt.ylabel("Distance in kms")
plt.title("Compare")
plt.legend()
plt.show()
```

```python
## Example for histogram
age = [22,24,26,28,27,30,84,90,65]
gaps = [0,10,20,30,40,50,60,70,80]
plt.hist(age,gaps,rwidth=0.6)
```

```python
## Example for Scatter plot
x = [1,1.5,2,2.5,2.8,3.6,4.9]
y = [5,5.5,6,6.6,7.3,8.1,10]

x1 = [7,7.5,8,8.5,10]
y1 = [3,3.5,4,4.5,5.6]

plt.scatter(x,y,label='blue dots',color='b')
plt.scatter(x1,y1,label = "red dots",color='r')
plt.legend()
plt.show()
```

```python
#Example for Area plot
days = [1,2,3,4,5,6,7]
sleep = [7,8,6,10,9,5,4]
playing = [3,2,5,7,4,3,2]
working = [10,9,6,12,5,13,7]

plt.plot([],[],color='c',label='sleep',linewidth=5)
plt.plot([],[],color='g',label='playing',linewidth=5)
plt.plot([],[],color='m',label='working',linewidth=5)

plt.stackplot(days, sleep,playing,working,colors=['c','g','m'])
plt.legend()
```

```python
## Example for pie chart
eat = [1,2,3,4,5,6,7]
sleep = [7,8,6,10,9,5,4]
playing = [3,2,5,7,4,3,2]
working = [10,9,6,12,5,13,7]
slices = [3,7,2,13]
activity = ['eat','sleep','playing','working']
cols = ['c','m','g','b']
plt.pie(slices,labels=activity,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.show()
```
